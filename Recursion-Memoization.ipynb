# Memoization for Recursion

```python
def fibonacci(n):
    if n==1:
        return 1
    elif n==2:
        return 1
    elif n>2:
        return fibonacci(n-1)+fibonacci(n-2)
for n in range(1,35):
    print(f"{n} -> {fibonacci(n)}",end=" ")
```

```python
# Improved Recursive Fibonacci
cache=[]
for i in range(1000):
    cache.append(0)
def fibonacci(n):
    if cache[n]!=0:
        return cache[n]
    if n==1:
        return 1
    elif n==2:
        return 1
    elif n>2:
        element=fibonacci(n-1)+fibonacci(n-2)
        cache[n]=element
        return element
```

```python
for n in range(1,1000):
    print(f"{n} -> {fibonacci(n)}",end=" ")
```

```python
from functools import lru_cache
@lru_cache(maxsize=1000)
def fibonacci(n):
    if n==0:
        return 1
    elif n==1:
        return 1
    elif n>1:
        return fibonacci(n-1)+fibonacci(n-2)
for n in range(1,100):
    print(f"{n} -> {fibonacci(n)}",end=" ")
```

```python
for i in range(20):
    print(fibonacci(i+1)/fibonacci(i))
```

```python
def index_fib(n):
    for i in range(len(cache)):
        if cache[i]==n:
            index=i
            break
    else:
        print("Not Found")
    return index   
            
```

```python
def index_fib(n,index):
    element=fibonacci(index)
    if n<element:
        return -1
    elif n==element:
        return index
    else:
        return index_fib(n,index+1)
```

```python
def pascalrow(n):
    if n==1:
        return [1]
    elif n>1:
        line=[1]
        prevline=pascalrow(n-1)
        for i in range(len(prevline)-1):
            line.append(prevline[i]+prevline[i+1])
        line.append(1)
    return line
        
    
```

```python
pascalrow(5)
```

```python
for i in range(1,5):
    print(pascalrow(i))
```
